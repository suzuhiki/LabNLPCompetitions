```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np 

%matplotlib inline
pd.options.display.max_columns = 32
pd.options.display.max_rows = 200
```

```python
PROJECT_ROOT_PATH = "../../"

datapath = PROJECT_ROOT_PATH + "lab_competition/data/"
outpath = PROJECT_ROOT_PATH + "lab_competition/output/01/"

import numpy as np
import collections
```

```python
import re

# 各データを読み込みリストに格納
def read_file(path):
    with open(path, mode="r") as f:
        result = f.read().splitlines()
    return result

train_text = read_file(datapath + "text.train.txt")
dev_text = read_file(datapath + "text.dev.txt")
test_text = read_file(datapath + "text.test.txt")
train_label = np.loadtxt(datapath + "label.train.txt")
dev_label = np.loadtxt(datapath + "label.dev.txt")
```

# ラベル作成

```python
def negaposi_3class(label_list):
    result = []
    for l in label_list:
        if l >= 1:
            result.append(1)
        elif l <= -1:
            result.append(-1)
        else:
            result.append(0)
    return result

def negative_2class(label_list):
    result = []
    for l in label_list:
        if l == -2:
            result.append(-2)
        elif l == -1:
            result.append(-1)
        else:
            result.append(0)
    return result

def positive_2class(label_list):
    result = []
    for l in label_list:
        if l == 2:
            result.append(2)
        elif l == 1:
            result.append(1)
        else:
            result.append(0)
    return result
```

```python
train_df = pd.DataFrame()
train_df["data"] = train_label
train_df["3class"] = negaposi_3class(train_label)
train_df["positive"] = positive_2class(train_label)
train_df["negative"] = negative_2class(train_label)
train_df["text"] = train_text

print(train_df.head(50))
```

```python
train_df.to_csv(datapath+"train_3class.csv", columns=["3class", "text"], index = False)
train_df[train_df["3class"] == 1].to_csv(datapath+"train_positive.csv", columns=["positive", "text"], index = False)
train_df[train_df["3class"] == -1].to_csv(datapath+"train_negative.csv", columns=["negative", "text"], index = False)
```

### devデータ

```python
dev_df = pd.DataFrame()
dev_df["data"] = dev_label
dev_df["3class"] = negaposi_3class(dev_label)
dev_df["positive"] = positive_2class(dev_label)
dev_df["negative"] = negative_2class(dev_label)
dev_df["text"] = dev_text

print(dev_df.head(50))
```
