```python
PROJECT_ROOT_PATH = "../../"

datapath = PROJECT_ROOT_PATH + "lab_competition/data/"
outpath = PROJECT_ROOT_PATH + "lab_competition/output/01/leaderboad/"
```

```python
import numpy as np

stepest_v4 = np.loadtxt(outpath + "v0.0.4_StepEstimator.txt")
lgb_v3 = np.loadtxt(outpath + "v0.0.3_LightGBM.txt")
random_forest_v2 = np.loadtxt(outpath + "v0.0.2_RandomForest.txt")
logisticR_v1 = np.loadtxt(outpath + "v0.0.1_LogisticRegression.txt")
linearSVC_v1 = np.loadtxt(outpath + "v0.0.1_LinearSVC.txt")
```

```python
import pandas as pd
```

```python
data_df = pd.DataFrame()

data_df["stepest_v4"] = stepest_v4
data_df["lgb_v3"] = lgb_v3
data_df["random_forest_v2"] = random_forest_v2
data_df["logisticR_v1"] = logisticR_v1
data_df["linearSVC_v1"] = linearSVC_v1
```

```python
data_df.describe()
```

```python
result_df = pd.DataFrame(columns=["-2", "-1", "0", "1", "2"], index=range(2500))
result_df.fillna(0, inplace=True)

for index, row in data_df.iterrows():
    for r in row:
        if r == -2.0:
            result_df.iat[index, 0] += 1
        elif r == -1.0:
            result_df.iat[index, 1] += 1
        elif r==0.0:
            result_df.iat[index, 2] += 1
        elif r==1.0:
            result_df.iat[index, 3] += 1
        elif r==2.0:
            result_df.iat[index, 4] += 1
    row["stepest_v4"] += 1
print(result_df)
```

```python
def cal_ensemble(df):
    result = []
    for index, row in df.iterrows():
        result.append(row.idxmax())
    return result

result = cal_ensemble(result_df)
```

```python
print(result)
```

```python
with open(outpath + "v0.0.2_EnsembleStacking.txt",'w') as f:
    f.writelines([d+"\n" for d in result])
```
